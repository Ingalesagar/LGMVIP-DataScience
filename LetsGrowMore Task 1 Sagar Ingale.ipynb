```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings('ignore')
%matplotlib inline
```

### Reading the Dataset

```python
df=pd.read_csv('iris.csv')
df
```

```python
df.rename(columns={'SepalLengthCm':'SepalLength', 'SepalWidthCm':'SepalWidth','PetalLengthCm':'PetalLength','PetalWidthCm':'PetalWidth'},inplace=True)
# renaming the column name SepalLengthCm as SepalLength , SepalWidthCm as SepalWidth , PetalLengthCm as PetalLength , PetalWidthCm as PetalWidth
```

```python
df['Species'].replace({'Iris-virginica':'virginica','Iris-setosa':'setosa','Iris-versicolor':'versicolor'},inplace=True)
# Replacing the Iris-virginica,Iris-setosa and Iris-versicolor by virginica,setosa and versicolor respectively 
```

```python
df= df.drop("Id", axis=1)
# As the Id column in the iris dataset resembles to the serial number of the rows inserted as it is not necessary for prediction 
# Droping the Id column from the data set
```

### Spliting the dataset as independent variable X and dependent variable y

```python
X=df.iloc[:,[0,1,2,3]]
y=df.Species
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=18)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

### Support Vector Machine

```python
from sklearn.svm import SVC  
from sklearn.metrics import classification_report, confusion_matrix 
svclassifier = SVC(kernel='poly', degree=8, gamma='auto')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))  
print(classification_report(y_test, y_pred))
```
